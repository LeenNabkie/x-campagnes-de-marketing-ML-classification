```
# Import libraries
## Basic libs
import pandas as pd
import numpy as np
import warnings
import io

## Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Configure libraries
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (10, 10)
plt.style.use('seaborn')

# Scalling library:
from sklearn.preprocessing import StandardScaler
# Coding library:
from sklearn.preprocessing import OneHotEncoder

#Decision tree model:
from sklearn import tree

#Random Forest Model:
from sklearn.ensemble import RandomForestClassifier

#GradientBoosting Model:
from sklearn.ensemble import GradientBoostingClassifier

```

```
## Loading data
from google.colab import files
uploaded = files.upload()

df_banque = pd.read_csv(io.BytesIO(uploaded['30636_A2021_Train_Valid.csv']))
df_banque.columns = df_banque.columns.str.replace(' ', '')

print(df_banque)
```

```
# Drop 'duration' column
df_banque = df_banque.drop('duration', axis=1)

#print(df_banque.info())
df_banque.head()
```

```
df_banque['subscribed'].value_counts()
```

```
# Missing value:
df_banque.isnull().sum()
```

```
# Scale Numeric Data
# Make a copy of our df_banque
df_scalled = df_banque.copy()

scaler = StandardScaler()
num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']
df_scalled[num_cols] = scaler.fit_transform(df_scalled[num_cols])

df_scalled.head()
```

```
# Encode our label column (category) by replacing 'yes' and 'no' with 1 and 0 respectively:
encoder = OneHotEncoder(sparse=False)
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

# Encode Categorical Data
df_encoded = pd.DataFrame(encoder.fit_transform(df_scalled[cat_cols]))
df_encoded.columns = encoder.get_feature_names(cat_cols)

# Replace Categotical Data with Encoded Data
df_scalled = df_scalled.drop(cat_cols ,axis=1)
df = pd.concat([df_encoded, df_scalled], axis=1)
# Encode target value
df['subscribed'] = df['subscribed'].apply(lambda x: 1 if x == 'yes' else 0)

df.head()
```

# Split Dataset for Training and Testing

```
# Select Features
feature = df.drop('subscribed', axis=1)

# Select Target
target = df['subscribed']

# Set Training and Testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature , target, shuffle = True, test_size=0.2, random_state=1)
```

```
print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of training label:', y_test.shape)
```

####Building Modeles:

# 1- Decision tree:

```
dtc = tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
```

# 2-Random Forest:

```
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
```

# 3- GradientBoostingClassifier:

```
gb_clf = GradientBoostingClassifier(n_estimators=20, learning_rate=1, max_features=2, max_depth=2, random_state=0)
gb_clf.fit(X_train, y_train)
```

#Evaluate the model's performance by using the following metrics:

# Accuracy, Precision, Recall, confussion matrix

```
def evaluate_model(model, x_test, y_test):
    from sklearn import metrics

    # Predict Test Data 
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'prec': prec, 'rec': rec, 'cm': cm}
```

# Evaluate  Decision tree Model:

```
dtc_eval = evaluate_model(dtc, X_test, y_test)

print('Accuracy of dtc:', dtc_eval['acc'])
print('Precision of dtc:', dtc_eval['prec'])
print('Recall of dtc:', dtc_eval['rec'])
print('Confusion Matrix of dtc:\n', dtc_eval['cm'])
```

# Evalute Random Forest Model:

```
rf_eval = evaluate_model(rf, X_test, y_test)

print('Accuracy of rf:', rf_eval['acc'])
print('Precision of rf:', rf_eval['prec'])
print('Recall of rf:', rf_eval['rec'])
print('Confusion Matrix of rf:\n', rf_eval['cm'])
```

# Evaluate GradientBoostingClassifier Model:

```
gb_clf_eval = evaluate_model(gb_clf, X_test, y_test)

print('Accuracy of gb_clf:', gb_clf_eval['acc'])
print('Precision of gb_clf:', gb_clf_eval['prec'])
print('Recall of gb_clf:', gb_clf_eval['rec'])
print('Confusion Matrix of gb_clf:\n', gb_clf_eval['cm'])
```

# Prediction:

```
df['subscribed_prediction'] = dtc.predict(feature)
df['subscribed_prediction'] = df['subscribed_prediction'].apply(lambda x: 'yes' if x==0 else 'no')

# Save new dataframe into csv file
df.to_csv('subscribed_prediction.csv', index=False)

df.head(10)
```
